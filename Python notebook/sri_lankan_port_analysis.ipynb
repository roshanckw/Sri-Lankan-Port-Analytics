| **import_dry_bulk** | Total import volume (in metric tons) of all dry bulk carriers entering the port at this date. |
| **import_general_cargo** | Total import volume (in metric tons) of all general cargo ships entering the port at this date. |
| **import_roro** | Total import volume (in metric tons) of all ro-ro ships entering the port at this date. |
| **import_tanker** | Total import volume (in metric tons) of all tankers entering the port at this date. |
| **import_cargo** | Total import volume (in metric tons) of all ships (excluding tankers) entering the port at this date. This is the sum of import_container, import_dry_bulk, import_general_cargo and import_roro. |
| **import** | Total import volume (in metric tons) of all ships entering the port at this date. This is the sum of import_container, import_dry_bulk, import_general_cargo, import_roro and import_tanker. |
| **export_container** | Total export volume (in metric tons) of all container ships entering the port at this date. |
| **export_dry_bulk** | Total export volume (in metric tons) of all dry bulk carriers entering the port at this date. |
| **export_general_cargo** | Total export volume (in metric tons) of all general cargo ships entering the port at this date. |
| **export_roro** | Total export volume (in metric tons) of all ro-ro ships entering the port at this date. |
| **export_tanker** | Total export volume (in metric tons) of all tankers entering the port at this date. |
| **export_cargo** | Total export volume (in metric tons) of all ships (excluding tankers) entering the port at this date. This is the sum of export_container, export_dry_bulk, export_general_cargo and export_roro. |
| **export** | Total export volume (in metric tons) of all ships entering the port at this date. This is the sum of export_container, export_dry_bulk, export_general_cargo, export_roro and export_tanker. |

## **2. Trade analysis**

**Business Objectives**
- Which ports handle the largest trade volumne?
- Which ports are import-oriented or export-oriented?
- What is the trade balance of each port?
- Which ports cotribute the most to Sri Lanka's maritime trade?

**Step 2.1 Create New Features**

```python
import numpy as np
```

```python
# Total Trade Volume
df["total_trade"] = df["import"] + df["export"]

# Trade Balance
# Positive -> More Exports
# Negative -> More Imports
df["trade_balance"] = df["export"] - df["import"]

# Import Percentage
df["import_percentage"] = np.where(
    df["total_trade"] > 0,
    (df["import"] / df["total_trade"]) * 100,
    0
)

# Export Percentage
df["export_percentage"] = np.where(
    df["total_trade"] > 0,
    (df["export"] / df["total_trade"]) * 100,
    0
)
```

```python
df[[
    "portname",
    "import",
    "export",
    "total_trade",
    "trade_balance",
    "import_percentage",
    "export_percentage"
]].head()
```

## **9.Operational KPI Analysis**

**Step 9.1 Create KPI Features**

To avoid division-by-zero errors, use np.where()

**Average Import per Ship**

On average, how much import corgo does each arriving ship carry?

```python
df["avg_import_per_ship"] = np.where(
    df["portcalls"] > 0,
    df["import"] / df["portcalls"],
    0
)
```

**Average Export per ship**

```python
df["avg_export_per_ship"] = np.where(
    df["portcalls"] > 0,
    df["export"] / df["portcalls"],
    0
)
```

**Average Total Trade per ship**

```python
df["avg_trade_per_ship"] = np.where(
    df["portcalls"] > 0,
    df["total_trade"] / df["portcalls"],
    0
)
```

**Import-Export Ration**

- 1 - Import oriented
- <1 - Export oriented

```python
df["import_export_ratio"] = np.where(
    df["export"] > 0,
    df["import"] / df["export"],
    np.nan
)
```

I prefer np.nan over +1 because it avoids artifically changing the ratio when exports are zero.

**Container ship share**

```python
df["container_share"] = np.where(
    df["portcalls"] > 0,
    df["portcalls_container"] / df["portcalls"],
    0
)
```

**Container Ship Share**

```python
df["container_share"] = np.where(
    df["portcalls"] > 0,
    df["portcalls_container"] / df["portcalls"],
    0
)
```

**Tanker Share**

```python
df["tanker_share"] = np.where(
    df["portcalls"] > 0,
    df["portcalls_tanker"] / df["portcalls"],
    0
)
```

**Dry Bulk Share**

```python
df["dry_bulk_share"] = np.where(
    df["portcalls"] > 0,
    df["portcalls_dry_bulk"] / df["portcalls"],
    0
)
```

**General Cargo Share**

```python
df["general_cargo_share"] = np.where(
    df["portcalls"] > 0,
    df["portcalls_general_cargo"] / df["portcalls"],
    0
)
```

**Ro-Ro Share**

```python
df["roro_share"] = np.where(
    df["portcalls"] > 0,
    df["portcalls_roro"] / df["portcalls"],
    0
)
```

### Engineered Features Dictionary

| :--- | :--- |
